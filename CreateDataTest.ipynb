```python
# Tạo bảng users
users = [(1, "alice", "alice@mail.com"), (2, "bob", "bob@mail.com")]
df_users = spark.createDataFrame(users, ["user_id", "name", "email"])
df_users.write.mode("overwrite").saveAsTable("soa_catalog.ecommerce.users")

# Tạo bảng products
products = [(201, "Áo thun", 150), (202, "Quần jean", 300), (203, "Giày", 500)]
df_prod = spark.createDataFrame(products, ["product_id", "name", "price"])
df_prod.write.mode("overwrite").saveAsTable("soa_catalog.ecommerce.products")

# Tạo bảng orders (rỗng)
empty_orders = spark.createDataFrame([], schema="order_id INT, user_id INT, product_id INT, quantity INT, amount INT, status STRING, timestamp STRING")
empty_orders.write.mode("overwrite").saveAsTable("soa_catalog.ecommerce.orders")

# Tạo bảng payments (rỗng)
empty_payments = spark.createDataFrame([], schema="payment_id INT, order_id INT, status STRING, timestamp STRING")
empty_payments.write.mode("overwrite").saveAsTable("soa_catalog.ecommerce.payments")

print("All tables created successfully!")
```
