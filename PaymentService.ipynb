```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

dbutils.widgets.text("order_id", "0")
dbutils.widgets.text("amount", "0")
order_id = int(dbutils.widgets.get("order_id"))
amount = int(dbutils.widgets.get("amount"))

# Giả lập thanh toán thành công
import datetime
payment_id = order_id * 100
new_payment = [(payment_id, order_id, "completed", datetime.datetime.now().isoformat())]

# Define schema to match table schema (INT for payment_id and order_id)
schema = StructType([
    StructField("payment_id", IntegerType(), True),
    StructField("order_id", IntegerType(), True),
    StructField("status", StringType(), True),
    StructField("timestamp", StringType(), True)
])

df_pay = spark.createDataFrame(new_payment, schema)
df_pay.write.mode("append").saveAsTable("soa_catalog.ecommerce.payments")
dbutils.notebook.exit("PAID")
```
