```python
from pyspark.sql.functions import sum, count

# 1. Tổng doanh thu
revenue_df = spark.sql("""
    SELECT SUM(amount) as total_revenue 
    FROM soa_catalog.ecommerce.orders 
    WHERE status = 'completed'
""")
revenue_df.show()

# 2. Top sản phẩm bán chạy
top_products = spark.sql("""
    SELECT product_id, COUNT(*) as order_count
    FROM soa_catalog.ecommerce.orders
    GROUP BY product_id
    ORDER BY order_count DESC
""")
top_products.show()

# 3. Lưu kết quả phân tích vào Delta table
revenue_df.write.mode("overwrite").saveAsTable("soa_catalog.ecommerce.revenue_summary")
top_products.write.mode("overwrite").saveAsTable("soa_catalog.ecommerce.top_products")

print("Analytics completed. Tables updated.")
```
